```python
import ffmpeg
import pandas as pd
from src.settings import ORKPJM_VIDEOS_DIR

pd.set_option('display.max_columns', 100)
```

```python
# uses ffprobe command to extract all possible metadata from the media file
videos_metadata = []
for vid_filepath in ORKPJM_VIDEOS_DIR.glob('*.mp4'):
    try:
        vid_probe = ffmpeg.probe(vid_filepath)
    except ffmpeg.Error as e:
        print(e.stderr)
        continue
    vid_metadata_record = vid_probe["format"]
    vid_metadata_record.update(vid_probe["streams"][0])
    videos_metadata.append(vid_metadata_record)

videos_metadata_df = pd.DataFrame.from_records(videos_metadata)
videos_metadata_df
```

```python
videos_metadata_df["duration"] = videos_metadata_df["duration"].astype(float)
videos_metadata_df.describe()
```
